```python
dfTweets = sqlContext.read.load('sample_twt2.jsonl', format='json')
dfTweets.printSchema()
```

```python
import pyspark.sql.functions as sf
hashtags = dfTweets.select('entities.hashtags.text').dropna()
hashtags = hashtags.filter(sf.size('text')>0)
hashtags = hashtags.select(sf.explode('text').alias('stuff'))
hashtags = hashtags.select(sf.lower(hashtags.stuff).alias('stuff'))
tophashtags = hashtags.groupBy('stuff').count().orderBy('count', ascending = False)
tophashtags.show()
```

```python
# toploc.write.csv('toplocations/toplocations.csv')
#toploc.write.format('csv').save('toplocations/toplocations.csv')
tophashtags.toPandas().to_csv('tophashtags/tophashtags.csv', encoding='utf-8', index=False)
```
