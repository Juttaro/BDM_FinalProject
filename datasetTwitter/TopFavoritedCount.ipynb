```python
dfTweets = sqlContext.read.load('stream_twt_may11.jsonl', format='json')
```

```python
dfTweets.printSchema()
```

### The tweets with the top favorite counts ==> ( screen_name, fav_count, tweet)

```python
df_favorite_user = dfTweets.select('retweeted_status.user.screen_name', dfTweets['retweeted_status.favorite_count'].alias('fav_count'), dfTweets['retweeted_status.text'].alias('tweet')).dropna()

df_favorite_user = df_favorite_user.dropDuplicates(['screen_name','tweet'])

df_favorite_user = df_favorite_user.select('screen_name', 'fav_count', 'tweet').orderBy('fav_count', ascending=False).limit(50)

# df_favorite_user = df_favorite_user.
#df_favorite_user = df_favorite_user.orderBy('fav_count', ascending = False)

df_favorite_user.show() #, truncate=False)


```
